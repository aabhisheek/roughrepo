```python
from textblob import TextBlob

# Example text
text = "I absolutely love this product! It's fantastic and exceeded my expectations."

# Perform sentiment analysis
analysis = TextBlob(text)

# Results
print("Sentiment Polarity:", analysis.sentiment.polarity)
print("Sentiment Subjectivity:", analysis.sentiment.subjectivity)

# Interpretation
if analysis.sentiment.polarity > 0:
    print("Positive Sentiment")
elif analysis.sentiment.polarity < 0:
    print("Negative Sentiment")
else:
    print("Neutral Sentiment")
```

```python
from transformers import pipeline

# Load sentiment-analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

# Example text
text = "I absolutely love this product! It's fantastic and exceeded my expectations."

# Perform sentiment analysis
result = sentiment_analyzer(text)

# Output
print(result)  # [{'label': 'POSITIVE', 'score': 0.9998}]
```

```python
import matplotlib.pyplot as plt

# Data for two graphs
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 15, 20, 25, 35]

# Plot both graphs
plt.plot(x1, y1, marker='o', label="Graph 1")
plt.plot(x2, y2, marker='s', label="Graph 2")

# Add title, labels, and legend
plt.title("Two Graphs on the Same Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show legend
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for two graphs
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]

x2 = [1, 2, 3, 4, 5]
y2 = [5, 15, 20, 25, 35]

# First subplot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.plot(x1, y1, marker='o', color='blue')
plt.title("Graph 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Second subplot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.plot(x2, y2, marker='s', color='green')
plt.title("Graph 2")
plt.xlabel("X-axis")

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for first graph
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]

# Data for second graph
x2 = [1, 2, 3, 4, 5]
y2 = [5, 15, 20, 25, 35]

# First figure
plt.figure(1)
plt.plot(x1, y1, marker='o', color='blue')
plt.title("Graph 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Second figure
plt.figure(2)
plt.plot(x2, y2, marker='s', color='green')
plt.title("Graph 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show both figures
plt.show()
```
